```python
import pandas as pd

#  Load PlayStore Apps Excel file
data_url = 'https://github.com/ine-rmotr-projects/project-files/files/4086772/playstore.xlsx'

# Read the `playstore.xlsx` Excel file from the given `data_url` and store it in a `playstore_df` DataFrame.
# playstore_df = pd.ExcelFile(data_url)
# playstore_df.sheet_names

playstore_df = pd.read_excel(data_url)
playstore_df.head(10)

```

```python
# - When reading in the file, only use the columns `'App', 'Rating', 'Installs', 'Rating', 'Genres', 'Last_Updated'`.
# results = playstore_df[['App', 'Rating',
#                     'Installs', 'Rating', 'Genres', 'Last_Updated']]
# results.head()
col_names = ['App', 'Rating',
                    'Installs', 'Rating', 'Genres', 'Last_Updated']


playstore_df = pd.read_excel(data_url,
                             usecols=col_names)
playstore_df.head(10)
```

```python
# - Make sure `Last_Updated` is in datetime format, try to do this while reading the file into the DataFrame.

playstore_df = pd.read_excel(data_url,
                             usecols=col_names,
                             parse_dates=['Last_Updated'])
playstore_df.head(10)
```

```python
# After reading the data, filter the records and keep only the top 25 with highest `Rating` (being 5 the highest possible rating value).
playstore_df= playstore_df.sort_values('Rating', ascending=False)
playstore_df.head(25)
```

```python
playstore_df[['Genres', 'Rating']].head(5).plot(kind='bar',figsize=(17, 5))
```
