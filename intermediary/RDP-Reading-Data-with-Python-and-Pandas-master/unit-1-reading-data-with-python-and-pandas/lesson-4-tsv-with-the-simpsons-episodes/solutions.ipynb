```python
# Read and store in a `simpsons` DataFrame the data within the `simpsons-episodes.tsv` TSV (_Tabular Separated Values_) file. This file contains information about all The Simpsons episodes.
# - Use correct separator as data is tabular separated.
import pandas as pd

simpsons = pd.read_csv('files/simpsons-episodes.tsv', sep='\t')
simpsons.head(10)



```

```python
# - Use the following `col_names` list as column names.
col_names = ['Title', 'Air date', 'Production code', 'Season', 'Number in season',
             'Number in series', 'US viewers (million)', 'Views', 'IMDB rating']
simpsons = pd.read_csv('files/simpsons-episodes.tsv',
                       sep='\t', names=col_names)
simpsons.head(15)
```

```python
# - Load just `Title`, `Air date`, `Production code` and `IMDB rating`.
# simpsons[['Title', 'Air date', 'Production code', 'IMDB rating']]
simpsons = pd.read_csv('files/simpsons-episodes.tsv',
                       sep='\t', names=col_names,
                       usecols=['Title', 'Air date', 'Production code', 'IMDB rating'])
simpsons.head(10)

```

```python
# - Don't load the first empty columns.
simpsons = pd.read_csv('files/simpsons-episodes.tsv',
                       sep='\t', names=col_names,skiprows=[1,2,3])
simpsons
```

```python
# - Set `Production code` as index.
# simpsons = pd.read_csv('files/simpsons-episodes.tsv',
#                        index_col=2,
#                        sep='\t', names=col_names,
#                        skiprows=[1,2,3])

simpsons = pd.read_csv('files/simpsons-episodes.tsv',
                       index_col=['Production code'],
                       sep='\t', names=col_names,
                       skiprows=[1,2,3])
simpsons.head(10)
```

```python
# - Null values are encoded as `no_val` values, be careful with that when loading the data.
simpsons = pd.read_csv('files/simpsons-episodes.tsv',
                       index_col=['Production code'],
                       sep='\t', names=col_names,
                       skiprows=[1,2,3], na_values='no_val')
simpsons.head(10)
result = simpsons[['Title', 'IMDB rating']].value_counts()
result.head(10).plot(kind='pie', figsize=(6,6))
```

```python
# - Parse the `Air date` columns as Date.
simpsons = pd.read_csv('files/simpsons-episodes.tsv',
                       index_col=['Production code'],
                       sep='\t', names=col_names,
                       skiprows=[1,2,3], na_values='no_val',
                       parse_dates=['Air date'])
simpsons.head(10)
```
