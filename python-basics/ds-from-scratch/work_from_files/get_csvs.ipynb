```python
from matplotlib import pyplot as plt
import pandas as pd
import csv
from collections import Counter

with open('../data/data.csv') as csv_file:
    csv_reader = csv.DictReader(csv_file)
    # row =next(csv_reader)
    count = Counter()

    for responses in csv_reader:
        count.update(responses['LanguagesWorkedWith'].split(';'))

common = count.most_common(15)

# Unpack a tuple into separate list using *
langs, popularity = [list(e) for e in zip(*common)]

langs.reverse()
popularity.reverse()

plt.barh(langs, popularity)
plt.title("Programming languages by popularity")
plt.ylabel("Languages ")
plt.xlabel("Most used")
plt.savefig("langs.png")
```

```python
# Using pandas

data = pd.read_csv("../data/data.csv", index_col=['Responder_id'])
counter = Counter()

for resp in data['LanguagesWorkedWith']:
    counter.update(resp.split(';'))


# e[::-1] is use to reverse a list. This give a reverse order of the original list
languages, poplty = [e[::-1] for e in zip(*counter.most_common(15))]

plt.barh(languages, poplty, color='g')
plt.title("Programming Language by Popularity")
plt.ylabel("Popularity ")
plt.xlabel("Languages ")

```
