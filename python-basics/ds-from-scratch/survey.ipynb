```python
# Sum of digits

def sum_of_digits(n) ->int:
    nums = 0;

    if n < 0:
        n *=-1
    if n == 0:
        return 1

    while n !=0:
        nums +=1;
        n //=10;
    return nums;


result: int = sum_of_digits(44);
result
```

```python
# Gradient

def sum_of_squares(v) -> int:
    return sum(v_i ** 2 for v_i in  v)


print(sum_of_squares([4, 2, 1,5]))
```
